## Problem Description

The numerical describe a constant coastal release of nitrate in the ocean. The equation can be written as 

$$\frac{\partial c}{\partial t}+v_0c(\frac{\partial c}{\partial x}+\frac{\partial c}{\partial y})=D(\frac{\partial^2 c}{\partial x^2}+\frac{\partial^2 c}{\partial y^2})$$

However, the boundary condition of this question is too complicated for analytical solving. Simplifications have been made.

### Boundary Condition Simplification

Constant input of nitrate was replaced by an initial input of nitrate. For open ocean, the nitrate concentration will remain constant 0, which is Dirichlet boundary condition. For the coast, a mirrored field is implemented with a doubled initial concentration release, to guarantee Neumann boundary condition $\frac{\partial c}{\partial n}=0$.

Based on all the simplifications above, another simplification of infinite boundary is made. The boundary condition is given by 
$$c_{\infty}=0$$

### Initial Condition

Considering that the constant input of nitrate is complicated to describe and to solve numerically, the initial condition is now
$$c(t=0,x=0,y=0)=2c_0$$

### Example Question (on textbook)

$$\frac{\partial u}{\partial t}=k\Delta u=k(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}+\frac{\partial^2 u}{\partial z^2})$$

Initial Condition: $u(\vec{x},0)=\phi(\vec{x})$



### Example Solution

$$u(\vec{x},t)=\iiint{S_3(\vec{x}-\vec{x}',t)\phi(\vec{x}')}d\vec{x}'$$

In which, 

$$S_3(\vec{x},t)=\frac{1}{(4\pi k t)^{3/2}} e^{-(x^2+y^2+z^2)/4kt}$$

### Modified Solution (for our problem, diffusion only)

$$c(\vec{x},t)=\iint{S_2(\vec{x}-\vec{x}',t)\phi(\vec{x}')}d\vec{x}'$$

In which, 

$$S_2(\vec{x},t)=\frac{1}{4\pi k t} e^{-(x^2+y^2)/4Dt}$$

Recall the initial condition: 

$$\phi(\vec{x},0)=2c_0\delta(x,y)$$

The solution for our diffusion problem is given by 

$$c(\vec{x},t)=\frac{2c_0}{4\pi D t}e^{-(x^2+y^2)/4Dt}$$

### Solution for Advection-Diffusion Problem

With constant and uniform flow $\vec{u}=(0,v_0)$, the **final** solution for advection-diffusion equation is given by:

$$c(\vec{x},t)=\frac{2c_0}{4\pi D t}e^{-[x^2+(y-v_0t)^2]/4Dt}$$
